# Initialize

```python
import pandas as pd

def load_excel_data(file_path: str, skiprows = 0):
    # Load the Excel file, skipping the first three rows
    df = pd.read_excel(file_path, skiprows=skiprows)
    return df
```

## Importing bank datas from bank 1.

```python
file_names = {
    "file_names": [
        "e8cd26b84f8040458156a4d3098ef801_5_Invoice_List.xlsx",
        "a1d86c7c072944108ce3344e138f4fcf_3_Bank1_Statement2.xls"
    ]
}
```

```python
print(file_names["file_names"][1])
```

```python
file_name = file_names["file_names"][1]
data = load_excel_data(file_path=f"excels/{file_name}", skiprows=3)
print(data)
```

```python
data.head()
```

```python
# Filter the DataFrame
df_filtered = data.loc[data["Nature de l'opération"] == 'Virements reçus']
print(df_filtered)
```

```python
df_filtered['Bank'] = 'Banque Principale'
df_filtered.head()
```

```python
df_filtered['Date opération'] = pd.to_datetime(df_filtered['Date opération'])
df_filtered.sort_values(by='Date opération', inplace=True)
df_filtered['TransactionNumber'] = df_filtered.groupby('Date opération').cumcount() + 1
df_filtered['TransactionNumber'] = 'C' + df_filtered['Date opération'].dt.strftime('%y%m%d') + df_filtered['TransactionNumber'].apply(lambda x: f'{x:03}')

df_filtered.head()
```

```python
file_name = file_names["file_names"][0]
data = load_excel_data(file_path=f"excels/{file_name}")
```

```python
def filter_pairs(group):
    # Find the indices of rows in the group where 'Montant' has a negative counterpart
    indices_to_remove = group[group['Montant'].isin(-group['Montant'])].index
    # Drop these indices from the group
    return group.drop(indices_to_remove)
```

```python
data_filtered = data.groupby('Acheteur', group_keys=False).apply(filter_pairs).reset_index(drop=True)
```

## Import invoice list 

```python
def load_excel_data_without_header(file_path: str):
    df = pd.read_excel(file_path, header=None)
    return df
```

```python
file_names = {
    "file_names": [
        "5_Invoice_List.xlsx",
    ]
}
```

```python
file_name = file_names["file_names"][0]
data = load_excel_data_without_header(file_path=f"excels/{file_name}")
# Get the number of columns in the DataFrame
num_columns = data.shape[1]

# Generate column names based on the number of columns
column_names = [f'Column{i+1}' for i in range(num_columns)]

# Assign the generated column names to the DataFrame
data.columns = column_names
```

## Import Invoice status 

```python
file_names = {
    "file_names": [
        "6_Invoice_Status.xlsx",
    ]
}
```

```python
file_name = file_names["file_names"][0]
data = load_excel_data(file_path=f"excels/{file_name}")
```

```python
# Iterate over the invoice numbers in 'data'
for invoice_number in data['Numéro']:
    # Find the corresponding line in 'InvoicesData' and update the 'Status' field
    data.loc[data['InvoiceNumber'] == invoice_number, 'Status'] = 'Soldée'

# Print the 'InvoicesData' DataFrame to verify the updates
print(data)
```
